```python
import pandas as pd

patients = pd.read_csv('https://raw.githubusercontent.com/jncraton/exampledata/master/cancer.csv')
patients
```

```python
y = patients['diagnosis']
X = patients.iloc[:, 2:]
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5, random_state=0)
```

```python
clf = RandomForestClassifier(max_depth=4, class_weight={
    'B': 1.0,
    'M': 1.0,
}, random_state=0)

clf.fit(X_train, y_train)
plot_confusion_matrix(clf, X_test, y_test)
plt.title('Benign vs Malignant Cancer')
clf.score(X_test, y_test)
```

```python
accuracy_score(y_test.values, clf.predict(X_test))
```

```python
recall_score(y_test.values, clf.predict(X_test), pos_label="M")
```

```python
precision_score(y_test.values, clf.predict(X_test), pos_label="M")
```

```python
f1_score(y_test.values, clf.predict(X_test), pos_label="M")
```

```python
print(classification_report(y_test.values, clf.predict(X_test)))
```
