```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

houses = pd.read_csv("https://raw.githubusercontent.com/jncraton/exampledata/master/kc-house-data.csv")

reg = LinearRegression()
reg.fit(houses[['sqft_living']], houses['price'])

houses.plot.scatter(x='sqft_living', y='price')

reg.score(houses[['sqft_living']], houses['price'])
plt.plot(houses[['sqft_living']], reg.predict(houses[['sqft_living']]), c='black')
```
